```python
# generating frequency table
li = [10,20,10,30,1,2,1,1,1]
freq = dict()

for i in range(len(li)):
    x = li[i]
    freq[x] = freq.get(x,0) + 1
    #             ☝️  it try to get freq[x]
    #              if not available keep `0`
```

```python
print(freq)
```

```python
nums = [1,2,3,4]
n = len(nums)
```

```python
left = [1] * n

for i in range(1,n):
    left[i] = nums[i-1] * left[i-1]
print(left)
```

```python
right = [1] * n
for i in range(n-1,-1,-1):
    right[i] = nums[i+1] * right[i+1]
print(right)
```

```python
res = []
for i in range(n):
    res.append(left[i] * right[i])
print(res)
```
